```python
import altair as alt

import pandas as pd

patients_full = pd.read_csv("https://raw.githubusercontent.com/NHameleers/pythonprogrammingcoursefhml/master/Practicals/Practical11/Data/patient_data.csv")

# taking a smaller sample
patients = patients_full.sample(frac=.2, random_state=42)

patients

alt.Chart(patients).mark_point().encode(
    x='age',
)
```
